```python
import numpy as np

def initialize_params():
    W = np.random.randn(2, 1)
    b = 0
    return W, b

def forward(X, W, b):
    return np.dot(X, W) + b

def compute_loss(y, y_hat):
    return np.mean((y - y_hat) ** 2)

def sgd(W, dW, lr):
    return W - lr * dW

def momentum(W, dW, v, lr, beta=0.9):
    v = beta * v + (1 - beta) * dW
    return W - lr * v, v

def adam(W, dW, m, v, t, lr=0.01):
    beta1, beta2 = 0.9, 0.999
    m = beta1 * m + (1 - beta1) * dW
    v = beta2 * v + (1 - beta2) * (dW ** 2)
    m_hat = m / (1 - beta1 ** t)
    v_hat = v / (1 - beta2 ** t)
    return W - lr * m_hat / (np.sqrt(v_hat) + 1e-8), m, v
X = np.array([[1, 2]])
y = np.array([[5]])

W, b = initialize_params()
m = v = np.zeros_like(W)

for epoch in range(1, 101): 
    y_hat = forward(X, W, b)
    dW = -2 * X.T.dot(y - y_hat)
    W, m, v = adam(W, dW, m, v, epoch)

print("Trained Weights:", W)  
```

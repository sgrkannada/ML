```python
 import numpy as np
 def relu(x):
    return np.maximum(0, x)
np.random.seed(42)
W1 = np.random.randn(3, 4)
b1 = np.zeros((1, 4))
W2 = np.random.randn(4, 2)
b2 = np.zeros((1, 2))
X = np.random.randn(1, 3)
Z1 = np.dot(X, W1) + b1
A1 = relu(Z1)
Z2 = np.dot(A1, W2) + b2
print("Output of Neural Network:\n", Z2)
```
